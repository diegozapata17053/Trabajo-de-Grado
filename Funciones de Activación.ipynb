# Importar Librerías

```
%matplotlib inline
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
```

```
def plot_act(i=1.0, actfunc=lambda x: x):
    ws = np.arange(-0.5, 0.5, 0.05)
    bs = np.arange(-0.5, 0.5, 0.05)

    X, Y = np.meshgrid(ws, bs)

    os = np.array([actfunc(tf.constant(w*i + b)).eval(session=sess) \
                   for w,b in zip(np.ravel(X), np.ravel(Y))])

    Z = os.reshape(X.shape)

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, rstride=1, cstride=1)

#Inciamos una sesión
sess = tf.Session();
#Creamos un simple valor real de 3 entradas
i = tf.constant([1.0, 2.0, 3.0], shape=[1, 3])
#Creamos una matrix de pesos
w = tf.random_normal(shape=[3, 3])
#Creamos un vector de los biases(sesgos)
b = tf.random_normal(shape=[1, 3])
#Modelo de la Función de Activación
def func(x): return x
# tf.matmul multiplicara la entrada (i) de los tensores y el tensor de peso (w) luego suman el resultado con el tensor de sesgo (b)
act = func(tf.matmul(i, w) + b)
#Evaluar el tensor a una matriz numpy
act.eval(session=sess)

plot_act(1.0, func)
```

```
#Función Activación Sigmoid
plot_act(1, tf.nn.sigmoid)

act = tf.nn.sigmoid(tf.matmul(i, w) + b)
act.eval(session=sess)
```

```
#Función Activación ReLU
plot_act(1, tf.nn.relu)

act = tf.nn.relu(tf.matmul(i, w) + b)
act.eval(session=sess)
```

```
#Función Activación Leaky-ReLU
plot_act(1, tf.nn.leaky_relu)

act = tf.nn.leaky_relu(tf.matmul(i, w) + b)
act.eval(session=sess)
```

```
#Función Activación SeLU
plot_act(1, tf.nn.selu)

act = tf.nn.selu(tf.matmul(i, w) + b)
act.eval(session=sess)
```

```
act = tf.nn.softmax(tf.matmul(i, w,) + b)
```

```
act.eval(session=sess)
```

```
plot_act(3, tf.nn.softmax)
```
